```python
import pandas as pd
from numpy import corrcoef, cumsum
from sklearn.decomposition import PCA

index = pd.read_csv("djia_index.csv")
prices = pd.read_csv("close_prices.csv")
```

```python
index.head()
```

```python
prices.head()
```

```python
X = prices.ix[:, 1:]
clf = PCA(n_components=10)
clf.fit(X)
clf.explained_variance_ratio_
```

```python
enough = 0
for c in cumsum(clf.explained_variance_ratio_):
    enough += 1
    if (c > 0.9):
        break
enough
```

```python
first_component = clf.transform(X)[:, 0]
print(first_component)
len(first_component)
```

```python
corrcoef(first_component, index["^DJI"])
```

```python
import numpy as np
max_index = np.argmax(clf.components_[0])
prices.columns[1:][max_index]
```
